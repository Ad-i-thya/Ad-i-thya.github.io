```python
supstore.drop_duplicates(inplace=True)
```

```python
supstore
```

```python
supstore
```

```python
supstore['Order Date'] = pd.to_datetime(supstore["Order Date"], infer_datetime_format=True)
```

```python
supstore.info()
```

#### Create columns for month & year

```python
supstore['Year'] = supstore['Order Date'].dt.year
```

```python
supstore['Month'] = supstore['Order Date'].dt.month
```

#### Products which are most often bought together

```python
df = supstore1[supstore1['Order ID'].duplicated(keep=False)]
```

```python
df
```

```python
df['Grouped'] = df.groupby('Order ID')['Product Name'].transform(lambda x: '_'.join(x))
```

```python
df[['Order ID', 'Grouped']]
```

```python
df1 = df[['Order ID', 'Grouped']].drop_duplicates()
```

```python
df1
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df1['Grouped']:
    row_list = row.split('_')
    count.update(Counter(combinations(row_list,2)))

for key,value in count.most_common(10):
    print(key, value)
```
